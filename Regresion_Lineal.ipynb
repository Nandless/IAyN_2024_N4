```python

# Importar las librerías necesarias
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2

# Preprocesamiento de datos
train_data, test_data, train_labels, test_labels = train_test_split(
    horas_estudio, calificaciones, test_size=0.2, random_state=42
)

# Crear una instancia del modelo de regresión lineal y entrenarlo
model = LinearRegression()
model.fit(train_data, train_labels)

# Realizar predicciones con el conjunto de datos de prueba
predictions = model.predict(test_data)

# Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predictions)
print("Error cuadrático medio:", mse)

# Graficar el modelo de regresión lineal con las predicciones
plt.scatter(test_data, test_labels, color='green')
plt.plot(test_data, predictions, color='blue', linewidth=3)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión lineal')
plt.show()

```
